```python
import requests
from bs4 import BeautifulSoup
keyValue = {
    'User-Agent': 'Mozilla/5.0'
}
for i in range(1,11):
    html = requests.get('https://price.pcauto.com.cn/comment/sg2734/p'+str(i)+'.html', headers = keyValue)
    html.encoding = html.apparent_encoding
    htmlText = BeautifulSoup(html.text, 'lxml')
    youdian = htmlText.find_all('div', attrs = {'class': 'conLit youdian'})
    quedian = htmlText.find_all('div', attrs = {'class': 'conLit quedian'})
    print(str(i)+'-----------------------优点有：')
    for y in youdian:
        print(y.text)
    print(str(i)+'------------------------缺点有：')
    for q in quedian:
        print(q.text)
```
