```python
# 北京西城区 1到10月份房价趋势图
house = pd.read_csv('house_price.csv', encoding='gbk')
yuefen = house[house['区名'] == '西城区']['月份']  
fangjia = house[house['区名'] == '西城区']['房价']  

for a,b in zip(yuefen, fangjia):
    plt.text(a,b,b)  # a,b表示坐标，b表示值

plt.plot(yuefen, fangjia, 'r-o')
plt.show()
```
